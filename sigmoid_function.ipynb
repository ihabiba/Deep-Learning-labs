```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Input values to test
x_values = np.array([-10, -5, -1, 0, 1, 5, 10])

# Apply the sigmoid function to each
y_values = sigmoid(x_values)

# Display the results
for i in range(len(x_values)):
    print(f"x = {x_values[i]:>3} ---> sigmoid(x) = {y_values[i]:.4f}")
```

```python
# Generate smooth x values for plotting
x = np.linspace(-10, 10, 200)
y = sigmoid(x)

plt.figure(figsize=(7,4))
plt.plot(x, y, label='Sigmoid Curve')
plt.scatter(x_values, y_values, color='red', label='Given points')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```
